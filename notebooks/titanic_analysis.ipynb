### 3. Структура данных
Таблица содержит информацию о пассажирах Титаника:
1. Номер пассажира ('PassengerId')
2. Выживаемость ('Survived') 0 = нет, 1 = да
3. Класс билета ('Pclass')   1 = первый, 2 = второй, 3 = третий
4. Имя ('Name')
5. Пол ('Sex')
6. Возраст ('Age')
7. Количество братья/сёстры + супруги на борту('SibSp')
8. Количество родители + дети на борту('Parch')
9. Номер билета ('Ticket')
10. Стоимость билета ('Fare')
11. Номер комнаты проживания ('Cabin')
12. Порт посадки ('Embarked') C = Cherbourg, Q = Queenstown, S = Southampton
