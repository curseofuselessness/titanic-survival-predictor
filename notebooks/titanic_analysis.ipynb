## Обработка пропущенных значений

## Обработка пропусков

план:
1. пропуски в Age заполнить медианой
2. пропуски в Cabin - замена признаком CabinKnown(известна ли команата, 0 = нет, 1 = да)
3. пропуски в Embarked - замена модой

### Обработка пропусков в Age

Значение медиан

```python
age_medians = df.groupby(['Pclass', 'Sex'])['Age'].median().unstack()
print(age_medians)
```

Количество пассажиров в группах


```python
passengers_amounts = df.groupby(['Pclass', 'Sex']).size().unstack()
print(passengers_amounts)
```

Заполнение пропусков

```python
df['Age'] = df.groupby(['Pclass', 'Sex'])['Age'].transform(lambda x : x.fillna(x.median()))
print(df['Age'].isna().sum()) # проверка количества пропусков после обработки
```
